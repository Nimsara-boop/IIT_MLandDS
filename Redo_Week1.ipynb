```python
# removing rows (rows - Instances. columns - Features)
# using drop(index) function, we can drop a row at a particular index

# row indexing starts at 0, 1, 2, 3,.. etc.

data_frame.drop(4, inplace=True)
data_frame.head()
```

```python
# to remove multiple rows, we include a list of indices
# here, row numbers 0, 1 and 2 are removed

data_frame.drop([0, 1, 2], inplace=True)
data_frame.head()
```
