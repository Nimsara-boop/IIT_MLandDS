```python
import plotly.express as px
import pandas as pd
```

```python
data_frame = pd.read_csv('/content/drive/MyDrive/Colab_Notebooks/Prepared_Mall_Customers.csv')
data_frame.head()
```

1 - Univariate Data Visualisation

```python


# Construct the histogram plot for the Spending_Score histogram
Spending_Score_fig = px.histogram(data_frame, x='Spending_Score')

# Displaying constructed histogram plot
Spending_Score_fig.show()
```

```python
# Construct bar chart plot for Customer_Satisfaction
Customer_Satisfaction_fig = px.bar(data_frame, x = 'Customer_Satisfaction')

#Display constructed bar chart
Customer_Satisfaction_fig.show()
```

```python
#Spending_Score_bar = px.bar(x=data_frame['ID'], y=data_frame['Spending_Score'])
Spending_Score_bar = px.bar(data_frame = data_frame, x='ID', y ='Spending_Score', color='Customer_Satisfaction')
Spending_Score_bar.show()
```

Customer Satisfaction VS Gender(Sex) - Stacked Bar Charts / Bivariate Bar Plots

```python
data_frame = pd.read_csv('/content/drive/MyDrive/Colab_Notebooks/Prepared_Mall_Customers.csv')
```

```python
data_frame['Sex'] = data_frame['Sex'].map({1:'Female', 0:'Male'})
Customer_Satisfaction_VS_Gender = px.bar(data_frame = data_frame, x = 'Customer_Satisfaction', color='Sex')
Customer_Satisfaction_VS_Gender.show()
```

Clustered Bar Chart  - Age VS Customer Satisfaction

```python
Age_Vs_Count_coloredwith_Customer_Satisfaction = px.histogram(data_frame = data_frame, x='Age', color="Customer_Satisfaction" , barmode="group")
Age_Vs_Count_coloredwith_Customer_Satisfaction.show()
```

Multivariate Plots

```python
Age_VS_Salary_coloredwith_CustomerSatisfaction = px.scatter(data_frame = data_frame, x = 'Age', y='Salary', color="Customer_Satisfaction")
Age_VS_Salary_coloredwith_CustomerSatisfaction.show()
```

```python
Age_VS_Salary_coloredwith_CustomerSatisfaction = px.scatter_3d(data_frame = data_frame, x = 'Age', y='Salary', z='Spending_Score', color="Customer_Satisfaction")
Age_VS_Salary_coloredwith_CustomerSatisfaction.show()
```
