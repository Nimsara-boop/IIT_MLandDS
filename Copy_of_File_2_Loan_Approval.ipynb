```python
# Calculate error for K Values between 1 and 10
error=[]

import numpy as np
import matplotlib.pyplot as plt

#Calculating error for K Values between 1 and 40
for i in range(1, 40):
  knn2 = KNeighborsClassifier(n_neighbors=i)
  knn2.fit(X_cls_train, Y_cls_train)
  pred_i = knn2.predict(X_cls_test)
  error.append(np.mean(pred_i != Y_cls_test))

plt.figure(figsize=(12,6))
plt.plot(range(1, 40), error, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Error')
```

To model your coursework dataset with Naive Bayes, you first need to load a machine learning package, module and function, then instantiate the Naive Bayes modelling algorithm. Train it with your training data subset. Select the code block which helps you build a Naive Bayes classification model.

```python
import numpy as np
import pandas as pd
import plotly.express as px
from sklearn.naive_bayes import GaussianNB

# build the naive bayes model
# Initialize the Gaussian Naive Bayes classifier
nb = GaussianNB()
# Fit the Naive Bayes model to the training data
nb.fit(X_cls_train, Y_cls_train)
# Make predictions on the test data
Y_pred = nb.predict(X_cls_test)

# evaluate the naive bayes model
# Import the classification_report function for model evaluation
from sklearn.metrics import classification_report
# Print the classification report comparing actual and predicted values
print(classification_report(Y_cls_test, Y_pred))
```

Code Reuse Session 2 - Q24;

In your coursework, you are using eager Machine Learning classification algorithms that have a number of learnable parameters and hyperparameters.

To estimate the parameters and find the best combination of hyperparameters while building your model, you should use a nested cross-validation method. Select the code block; if used, it loads a suitable package, module, and function to perform an exhaustive hyperparameter search using cross-validation to find the best model configuration.

```python
# Logistic Regression - GridSearch CV

# Import GridSearchCV for hyperparameter tuning and LogisticRegression for the model
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

# Initialize Logistic Regression model with balanced class weight and increased max_iter
log = LogisticRegression(class_weight='balanced', max_iter=1000)

# Define the hyperparameter grid for Logistic Regression
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['lbfgs', 'liblinear'],
    'penalty': ['l2']
}

# Initialize GridSearchCV for Logistic Regression with 5-fold cross-validation and roc_auc scoring
log_gscv = GridSearchCV(log, param_grid, cv=5, scoring='roc_auc')
# Fit the GridSearchCV model to the training data
log_gscv.fit(X_cls_train, Y_cls_train)

# Print the best hyperparameters found by GridSearchCV
print("for the logistic regression model : ", log_gscv.best_params_)

# Make predictions on the test data using the best Logistic Regression model
y_pred_log = log_gscv.predict(X_cls_test)
# Print the classification report for Logistic Regression predictions
print(classification_report(Y_cls_test, y_pred_log))



# Naive Bayes - GridSearchCV

# Import Gaussian Naive Bayes classifier
from sklearn.naive_bayes import GaussianNB

# Initialize Gaussian Naive Bayes model
nb = GaussianNB()

# Define the hyperparameter grid for Naive Bayes (var_smoothing)
param_grid = {
    'var_smoothing': np.logspace(-12, 0, 50)
}

# Initialize GridSearchCV for Naive Bayes with 5-fold cross-validation and roc_auc scoring
nb_gscv = GridSearchCV(nb, param_grid, cv=5, scoring='roc_auc')
# Fit the GridSearchCV model to the training data
nb_gscv.fit(X_cls_train, Y_cls_train)

# Print the best hyperparameters found by GridSearchCV
print("for the naive bayes model ",nb_gscv.best_params_)

# Make predictions on the test data using the best Naive Bayes model
y_pred_nb = nb_gscv.predict(X_cls_test)
# Print the classification report for Naive Bayes predictions
print(classification_report(Y_cls_test, y_pred_nb))
```
